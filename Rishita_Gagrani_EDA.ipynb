# Data Science Assignment: eCommerce Transactions Dataset

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading datasets from the uploaded files
customers_path = '/content/Customers.csv'
products_path = '/content/Products.csv'
transactions_path = '/content/Transactions.csv'

# Reading the CSV files
customers = pd.read_csv(customers_path)
products = pd.read_csv(products_path)
transactions = pd.read_csv(transactions_path)

print("Data loaded successfully.")
```

```python
# Displaying dataset information
print("Customers Dataset Info:")
print(customers.info())
print("\nProducts Dataset Info:")
print(products.info())
print("\nTransactions Dataset Info:")
print(transactions.info())
```

```python
# Generating descriptive statistics for the datasets
print("Customers Dataset Statistics:\n", customers.describe(include='all'))
print("\nProducts Dataset Statistics:\n", products.describe(include='all'))
print("\nTransactions Dataset Statistics:\n", transactions.describe(include='all'))
```

```python
# Visualizing customer regions
data_region = customers['Region'].value_counts()
data_region.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Customer Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

```python
products['Category'].value_counts().plot(kind='bar', color='green', edgecolor='black')
plt.title('Product Categories Distribution')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(transactions['TotalValue'], bins=20, color='orange', edgecolor='black')
plt.title('Transaction Total Value Distribution')
plt.xlabel('Total Value')
plt.ylabel('Frequency')
plt.show()
```
